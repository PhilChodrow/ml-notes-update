## back to Xenocanto

```python
import requests

def query(species, max_length,key):
    query_url = f'https://xeno-canto.org/api/3/recordings?query=sp:"{species}"+len:"<{max_length}"&key={key}'
    response = requests.get(query_url)
    data = response.json()
    return data

species_list = ["dryobates pubescens", "melanerpes carolinus"]

for species in species_list:
    os.mkdir(f"../data/birdcall/{species}")
    q = query(species, 10, key)
    for recording in q["recordings"]:
        print(recording["file"])
        # download the file if the filename includes mp3
        if recording["file-name"].endswith(".mp3"):
            response = requests.get(recording["file"])
            with open(f"../data/birdcall/{species}/{recording['id']}.mp3", "wb") as file:
                file.write(response.content)
```
